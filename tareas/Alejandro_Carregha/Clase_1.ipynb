```python
#generar datos lineales
def generar_datos_lineales(
        n: int = 100,
        slope: float = 0.5,
        intercept: float = 1,
        noise_scale: float = 1.5
) -> pd.DataFrame:
    X = np.linspace(0,10,n)
    y = slope  *X + intercept + np.random.normal(scale = noise_scale, size = n)
    df = pd.DataFrame({"x":X, "y": y})
    return df
```

```python
df.y.quantile(q = [0.25, 0.35])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
lr = LinearRegression()
lr
```

```python
lr.fit(df[["x"]], df["y"])
```

```python
df["y_pred_lr"] = lr.predict(df[["x"]])
sns.scatterplot(data = df, x= "x", y = "y")
sns.lineplot(data = df, x = "x", y = "y_pred_lr")
```

```python
#KNN
#1. definir
knr = KNeighborsRegressor()
#2. ajustar
knr.fit(df[["x"]], df.y)
#3. predecir
knr.predict(df[["x"]])

y_pred_knr = knr.predict(df[["x"]])
```
